## Importing the libraries and open a image with opencv

```python
import cv2
img = cv2.imread("cat.jpg")
cv2.imshow("Cat Image",img)
cv2.waitKey(0)
```

```python
import cv2
from matplotlib import pyplot as plt
img = cv2.imread("cat.jpg")
# cv2.imshow("new",img)
# cv2.waitKey(0)
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB) ##Opencv follows BGR and matplotlib follows RGB
plt.imshow(img)
plt.xticks([]),plt.yticks([])
plt.show()
```

# Grayscaling the image

# Resizing The Image

```python
jerry = cv2.imread("jerry.jpg")
jerry = cv2.cvtColor(jerry,cv2.COLOR_BGR2RGB)
plt.imshow(jerry)
plt.show()
```

```python
print(jerry.shape)
```

```python
##let Change the width and height
width,height = 600,400
img_resized  = cv2.resize(jerry, (width,height))

plt.imshow(img_resized)
#plt.xticks([]), plt.yticks([]) (To hide the graph ticks)
plt.show()
```

# Cropping the image

```python
jerry= cv2.cvtColor(jerry,cv2.COLOR_BGR2RGB)
img_cropped = jerry[0:300,0:800]
cv2.imshow("Cat_crop", img_cropped)
cv2.waitKey(0)  #'''Using opencv to open in a separete window'''
```

```python
img_cropped = cv2.cvtColor(img_cropped,cv2.COLOR_BGR2RGB)
plt.imshow(img_cropped)
plt.xticks([]), plt.yticks([])
plt.show()
```

# Flipping the image

```python
img_flip = cv2.flip(img,0)   # 0 for flipping about x-axis (vertically)
 # 1 for flipping about y- axis (Horizontally)
cv2.imshow("Cat Imsge",img_flip)
cv2.waitKey(0)
```

```python
img_flip = cv2.flip(jerry,0)
img_flip = cv2.cvtColor(img_flip,cv2.COLOR_BGR2RGB)
plt.imshow(img_flip)
plt.show()
```

# Drawing Shapes

## 1) To draw line on a image

```python
 img_line = cv2.line(jerry, (0,0),(400,200),(55,20,255),5)  
```

1)-2nd and the 3rd arguments are the coordinates of the start and the end point of the line
2)-4th argument is color channel (B,G,R)
3)-5th argument is thickness of the line

```python
cv2.imshow("Line on image",img_line)
cv2.waitKey(0)
```

```python
img_line = cv2.cvtColor(img_line,cv2.COLOR_BGR2RGB)
plt.imshow(img_line)
plt.show()
```

## 2) To draw a rectangle

```python
jerry = cv2.imread("jerry.jpg")
img_rect = cv2.rectangle(jerry, (10,10),(500,400),(25,200,25),5)
img_rect = cv2.cvtColor(img_rect,cv2.COLOR_BGR2RGB)
plt.imshow(img_rect)
plt.show()
```

## 3)To draw a circle 

```python
jerry = cv2.imread("jerry.jpg")
img_cir =cv2.circle(jerry,(300,200),200,(0,50,250),10)
img_cir = cv2.cvtColor(img_cir,cv2.COLOR_BGR2RGB)
plt.imshow(img_cir)
plt.show()
```

```python
cv2.imshow("Circle on image",img_cir)
cv2.waitKey(0)
```

# Inserting a  text on a image

```python
font = cv2.FONT_HERSHEY_SIMPLEX    #SELECT A FOND FROM THE AVAILABLE FONTS
```

```python
jerry = cv2.imread("jerry.jpg")
img_txt = cv2.putText(jerry,"Hi I am Jerry",(0,500),font, 2 ,(255,5,100),8,cv2.LINE_AA)
```

3rd argument - Coordinate of the botton-Left corner of the text string ,
                  5th argument - font scale ,
                     6th argument - BGR ,
                   7th argument - Thickness ,
                 8th - line Type (optional) 


```python
cv2.imshow("Hi ",img_txt)
cv2.waitKey(0)
```

```python
img_txt = cv2.cvtColor(img_txt,cv2.COLOR_BGR2RGB)
plt.imshow(img_txt)
plt.show()
```
