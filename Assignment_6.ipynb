## Study and Implementation of Morphological Operations

```python
import cv2
from google.colab.patches import cv2_imshow
```

```python
import matplotlib.pyplot as plt
 
```

```python
img = cv2.imread('groot.jpg')
```

```python
img
```

```python
cv2_imshow(img)
```

###Erosion

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
eroded = cv2.erode(gray.copy(), kernel=(3,3), iterations = 4)
cv2_imshow(gray)
cv2_imshow(eroded)
 
```

###Dilation

```python
dilated = cv2.dilate(gray.copy(), kernel = (3,3), iterations = 4)
cv2_imshow(dilated)
cv2_imshow(gray)
```

##Opening

```python
kernelSize=(3,3)
kernel = cv2.getStructuringElement(cv2.MORPH_RECT, kernelSize)
opening = cv2.morphologyEx(gray.copy(), cv2.MORPH_OPEN, kernel)
cv2_imshow(opening)
cv2_imshow(gray)
```

##Closing

```python
closing = cv2.morphologyEx(gray.copy(), cv2.MORPH_OPEN, kernel)
cv2_imshow(closing)
cv2_imshow(gray)
```
