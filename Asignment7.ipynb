# Assignment: Study and Implement Harris Corner Point Detection.


```python
import numpy as np
import cv2
```

```python
img = cv2.imread('images.jpg')
```

```python
cv2_imshow(img)
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray)
```

```python
gray = np.float32(gray)
dest = cv2.cornerHarris(gray, 2, 5, 0.07)
dest = cv2.dilate(dest, None)
img[dest>0.005*dest.max()]=[0,0,255]
cv2_imshow(img)
```
