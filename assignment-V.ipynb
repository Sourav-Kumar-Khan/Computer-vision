```python
import cv2
from matplotlib import pyplot as plt
```

```python
image =cv2.imread('dog.jpg')
```

```python
img = cv2.resize(image,(300,200))
```

```python
from google.colab.patches import cv2_imshow
cv2_imshow(img)
```

```python
img= cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
cv2_imshow(img)
```

```python
_, th1 = cv2.threshold(img, 50, 255, cv2.THRESH_BINARY)
_, th2 = cv2.threshold(img, 50, 255, cv2.THRESH_BINARY_INV)
_, th3 = cv2.threshold(img, 127, 255, cv2.THRESH_TRUNC)
_, th4 = cv2.threshold(img, 127, 255, cv2.THRESH_TOZERO)
_, th5 = cv2.threshold(img, 127, 255, cv2.THRESH_TOZERO_INV)
_,th6 = cv2.threshold(img,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
```

####Titles for the images

```python
title=['Original Image','Binary Image','Binary Inverted Image','Truncated Image','Tozero', 'Tozero Inverted','Otsu\'s Thresholding ']
image=[img,th1,th2,th3,th4,th5,th6]
```

##Plotting the image side by side

```python
plt.figure(figsize=(12,8))
for i in range(7):
  plt.subplot(2,4,i+1)
  plt.imshow(image[i],'gray',vmin=0,vmax=255)
  plt.title(title[i])
  print('\n')
plt.show()
```
